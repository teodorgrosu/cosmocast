# SO + Planck noise curves

```python
import numpy as np
import matplotlib.pyplot as plt
from classy import Class
from itertools import combinations_with_replacement as cwr
from copy import deepcopy
import astropy.units as u

# Import the official SO noise calculator (public v3.1.2)
from so_noise_models.so_models_v3 import SO_Noise_Calculator_Public_v3_1_2 as so_models
```

## SO LAT noise curves

This section reproduces the original **LAT temperature** and **LAT polarization** noise plots.

```python
# mode = 0 is for threshold, mode = 1 is for baseline, mode = 2 is for goal
mode=1
fsky=0.4
ellmax=1e4
el=50.

lat = so_models.SOLatV3point1(mode, el=el)
corr_pairs = [(0,1),(2,3),(4,5)]

bands = lat.get_bands()
N_bands = len(bands)

ell, N_ell_LA_T_full, N_ell_LA_P_full = lat.get_noise_curves(
    fsky, ellmax, 1, full_covar=True, deconv_beam=True)

WN_levels = lat.get_white_noise(fsky)**.5

N_ell_LA_T  = N_ell_LA_T_full[range(N_bands), range(N_bands)]
N_ell_LA_Tx = [N_ell_LA_T_full[i, j] for i, j in corr_pairs]
N_ell_LA_P  = N_ell_LA_P_full[range(N_bands), range(N_bands)]
N_ell_LA_Px = [N_ell_LA_P_full[i, j] for i, j in corr_pairs]

def plot_temperature_noise_curves():
    ## plot the temperature noise curves
    plt.figure()
    for i in range(N_bands):
        plt.loglog(ell, N_ell_LA_T[i], label='%i GHz' % (bands[i]), ls='-', lw=2.)
    for _c, (i, j) in enumerate(corr_pairs):
        plt.loglog(ell, N_ell_LA_T_full[i, j],
                label=r'$%i \times %i$ GHz' % (bands[i], bands[j]), lw=1.5)

    plt.title(r"$N(\ell$) Temperature", fontsize=18)
    plt.ylabel(r"$N(\ell$) [$\mu$K${}^2$]", fontsize=16)
    plt.xlabel(r"$\ell$", fontsize=16)
    plt.ylim(5e-7, 100)
    plt.xlim(100, 10000)
    plt.legend(loc='lower left', ncol=2, fontsize=8)
    plt.grid()

plot_temperature_noise_curves()
```

## SO polarization noise curves

This reproduces the original LAT polarization plot and **adds SO SAT polarization curves** (auto only).

```python
# SO SAT (v3.1.2) polarization noise curves (auto only; SAT model has no T)
sat = so_models.SOSatV3point1(mode, one_over_f_mode=0)  # 0=pessimistic, 1=optimistic
sat_bands = sat.get_bands()
N_sat_bands = len(sat_bands)
ell_sat, _, N_ell_SA_P_full = sat.get_noise_curves(
    fsky, ellmax, 1, full_covar=True, deconv_beam=True
)
N_ell_SA_P = N_ell_SA_P_full[range(N_sat_bands), range(N_sat_bands)]

def plot_polarization_noise_curves_with_sat(include_LAT=False):
    ## plot the polarization noise curves (LAT) + SAT additions
    plt.figure()
    if include_LAT:
        for i in range(N_bands):
            plt.loglog(ell, N_ell_LA_P[i], label='%i GHz' % (bands[i]), ls='-', lw=2.)
        for _c, (i, j) in enumerate(corr_pairs):
            plt.loglog(ell, N_ell_LA_P_full[i, j],
                    label=r'$%i \times %i$ GHz' % (bands[i], bands[j]), lw=1.5)

    # SAT additions (auto only). Keep original styling intact; add dashed lines for SAT.
    for i in range(N_sat_bands):
        plt.loglog(ell_sat, N_ell_SA_P[i], label='SAT %i GHz' % (sat_bands[i]), ls='--', lw=2.)

    plt.title(r"$N(\ell$) Polarization", fontsize=18)
    plt.ylabel(r"$N(\ell$) [$\mu$K${}^2$]", fontsize=16)
    plt.xlabel(r"$\ell$", fontsize=16)
    plt.ylim(5e-7, 1e0)
    plt.xlim(30, 500)
    plt.legend(loc='upper left', ncol=2, fontsize=9)
    plt.grid()

plot_polarization_noise_curves_with_sat()
```

## Sanity check against Eq.(1) in arXiv:1808.07445 (original plot)

This section reproduces the original 93 GHz comparison plot.

```python
def noise_curve_93GHz_baseline(ell):

    arcmin = 0.000291 # rad
    Delta_T = 8.0*u.uK*arcmin
    theta_FWHM = 2.2*arcmin
    Nred = 1800*u.uK**2*u.s
    fsky = 0.4
    ell_knee = 1000
    alpha_knee = -3.5

    Nwhite = Delta_T.value**2*np.exp( ell*(ell+1)*theta_FWHM**2/(8*np.log(2)) ) # unit is uK^2
    
    Natm = (Nred*(4*np.pi*fsky)/(157680000*u.s*0.2*0.85) ).value # unit is uK^2
    Natm *= (ell/ell_knee)**alpha_knee
    Natm *= np.exp(ell*(ell+1)*theta_FWHM**2/(8*np.log(2)))

    Nell = Nwhite + Natm

    return np.array(Nell)

def noise_curve_93GHz_goal(ell):

    arcmin = 0.000291 # rad
    Delta_T = 5.8*u.uK*arcmin
    theta_FWHM = 2.2*arcmin
    Nred = 1800*u.uK**2*u.s
    fsky = 0.4
    ell_knee = 1000
    alpha_knee = -3.5

    Nwhite = Delta_T.value**2*np.exp( ell*(ell+1)*theta_FWHM**2/(8*np.log(2)) ) # unit is uK^2
    
    Natm = (Nred*(4*np.pi*fsky)/(157680000*u.s*0.2*0.85) ).value # unit is uK^2
    Natm *= (ell/ell_knee)**alpha_knee
    Natm *= np.exp(ell*(ell+1)*theta_FWHM**2/(8*np.log(2)))

    Nell = Nwhite + Natm

    return np.array(Nell)

def noise_curve_93GHz_test(ell):

    arcmin = 0.000291 # rad
    Delta_T = 8.0*u.uK*arcmin
    theta_FWHM = 2.2*arcmin
    # Nred = 1800*u.uK**2*u.s 
    Nred = 230/2*u.uK**2*u.s #factor of 2 is for FOV modification?
    fsky = 0.4
    ell_knee = 1000
    alpha_knee = -3.5

    Nwhite = Delta_T.value**2*np.exp( ell*(ell+1)*theta_FWHM**2/(8*np.log(2)) ) # unit is uK^2
    
    Natm = (Nred*(4*np.pi*fsky)/(157680000*u.s*0.2*0.85) ).value # unit is uK^2
    Natm *= (ell/ell_knee)**alpha_knee
    Natm *= np.exp(ell*(ell+1)*theta_FWHM**2/(8*np.log(2)))

    Nell = Nwhite + Natm

    return np.array(Nell)*u.uK**2
```

## Build noise dictionaries (original structure + SAT)

This reproduces the original `N_ell_T` / `N_ell_P` dictionaries and adds **SAT auto polarization** entries.

```python
frequencies = {
    "LAT": ["27", "39", "93", "145", "225", "280"],
    "SAT": ["27", "39", "93", "145", "225", "280"],
    "Planck": ["100", "143", "217", "353"]
}

f_pairs_LAT = ["LAT_27xLAT_27",
               "LAT_39xLAT_39",
               "LAT_93xLAT_93",
               "LAT_145xLAT_145",
               "LAT_225xLAT_225",
               "LAT_280xLAT_280",
               "LAT_27xLAT_39",
               "LAT_93xLAT_145",
               "LAT_225xLAT_280"]

f_pairs_Planck = ["Planck_{}xPlanck_{}".format(*cross) for cross in cwr(frequencies["Planck"], 2)]
```

```python
N_ell_T = {"LAT_{}xLAT_{}".format(*cross): ell * 0. for cross in cwr(frequencies["LAT"], 2)}
N_ell_P = deepcopy(N_ell_T)

# Fill LAT (autos + selected correlated pairs)
N_ell_T_list = np.concatenate([N_ell_LA_T, N_ell_LA_Tx])
N_ell_P_list = np.concatenate([N_ell_LA_P, N_ell_LA_Px])

for i, f_pair in enumerate(f_pairs_LAT):
    N_ell_T[f_pair] = N_ell_T_list[i]
    N_ell_P[f_pair] = N_ell_P_list[i]

# Add SAT polarization autos
for i, nu in enumerate([str(int(x)) for x in sat_bands]):
    key = f"SAT_{nu}xSAT_{nu}"
    N_ell_P[key] = N_ell_SA_P[i]
```

```python
sigma = {"Planck_{}xPlanck_{}".format(*cross): 0. for cross in cwr(frequencies["Planck"], 2)}
sigma_pol = deepcopy(sigma)

# converted temperature noise level from Table 4 of 1807.06205 to micro-Kelvin*arcmin
sigma["Planck_100xPlanck_100"] = 77.4
sigma["Planck_143xPlanck_143"] = 33.0
sigma["Planck_217xPlanck_217"] = 46.80
sigma["Planck_353xPlanck_353"] = 153.6

# converted polarization noise level from Table 4 of 1807.06205 to micro-Kelvin*arcmin
sigma_pol["Planck_100xPlanck_100"] = 117.6
sigma_pol["Planck_143xPlanck_143"] = 70.2
sigma_pol["Planck_217xPlanck_217"] = 105.0
sigma_pol["Planck_353xPlanck_353"] = 438.6

for f_pair in f_pairs_Planck:
    sigma_rad = np.deg2rad(sigma[f_pair]) / 60
    N_ell_T[f_pair] = ell * 0 + sigma_rad**2
    sigma_pol_rad = np.deg2rad(sigma_pol[f_pair]) / 60
    N_ell_P[f_pair] = ell * 0 + sigma_pol_rad**2
```

```python
import re
import matplotlib.pyplot as plt

_planck_key_re = re.compile(r"^Planck_(\d+)xPlanck_(\d+)$")

def plot_planck_noise_curves(N_ell_dict, ell, which="T", autos_only=True):
    """
    Plot Planck noise curves from a dict keyed like:
      Planck_100xPlanck_100, Planck_100xPlanck_143, ...

    autos_only=True plots only autos (100x100, 143x143, ...)
    """
    keys = []
    for k in N_ell_dict.keys():
        m = _planck_key_re.match(k)
        if not m:
            continue
        nu1, nu2 = int(m.group(1)), int(m.group(2))
        if autos_only and nu1 != nu2:
            continue
        keys.append((nu1, nu2, k))

    keys.sort(key=lambda t: (t[0], t[1]))

    plt.figure()
    for nu1, nu2, k in keys:
        lab = f"{nu1} GHz" if nu1 == nu2 else f"{nu1}×{nu2} GHz"
        plt.loglog(ell, N_ell_dict[k], label=lab)

    plt.title(rf"$N(\ell$) Planck {which}", fontsize=18)
    plt.ylabel(r"$N(\ell$) [$\mu$K${}^2$]", fontsize=16)
    plt.xlabel(r"$\ell$", fontsize=16)
    plt.ylim(5e-7, 1e0)
    plt.xlim(2, 10000)
    plt.legend(loc="best", ncol=2, fontsize=9)
    plt.grid()

# Examples:
plot_planck_noise_curves(N_ell_T, ell, which="T", autos_only=True)
plot_planck_noise_curves(N_ell_P, ell, which="P", autos_only=True)

# If you *also* want cross-frequency noise curves (mostly zero in your notebook):
# plot_planck_noise_curves(N_ell_T, ell, which="T", autos_only=False)
```
